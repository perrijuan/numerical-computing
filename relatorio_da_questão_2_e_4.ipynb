## Metodologia

### 1. Implementação da Soma de Kahan

A soma de Kahan é uma técnica utilizada para melhorar a precisão na soma de números em ponto flutuante, corrigindo os erros de arredondamento que se acumulam ao longo de somas sucessivas. Essa técnica é especialmente importante em séries numéricas onde pequenas perdas de precisão podem se acumular e causar erros significativos.

#### Algoritmo de Soma de Kahan

O algoritmo foi implementado da seguinte maneira:

```python
def soma_kahan(n, valor):
  soma = 0.0
  e = 0.0  # Variável de compensação

  for _ in range(n):
      y = valor - e  # Corrige o valor pelo erro anterior
      t = soma + y   # Adiciona o valor corrigido à soma
      e = (t - soma) - y  # Atualiza o erro
      soma = t

  return soma
```

### 2. Comparação com a Soma Convencional

Foi implementada uma função para realizar a soma convencional de números em ponto flutuante, para compará-la com a soma de Kahan:

```python
def floaterro(n, tamanho):
  soma = 0
  for _ in range(n):
    soma += tamanho
  return soma
```

### 3. Comparação dos Resultados

Para comparar os resultados, utilizamos valores de entrada variados, como `n = 20, 50, 100 e 1000`, e medimos a diferença entre as somas utilizando a função:

```python
def comparando_valores(n, entrada):
  somatorio = floaterro(n, int(entrada))
  metodo_kahan = soma_kahan(n, entrada)
  diferença_valor = somatorio - metodo_kahan
  return diferença_valor
```

## Resultados

Os resultados demonstraram que, para valores menores de `n`, a diferença entre os métodos é mínima. No entanto, à medida que o valor de `n` aumenta, a precisão do método de Kahan se destaca, apresentando uma diferença considerável em relação à soma convencional.

Por exemplo, para `n > 1000`, observamos que a soma convencional começa a apresentar um erro de arredondamento significativo, enquanto o método de Kahan corrige essas imprecisões.

## Discussão

A melhoria observada com a soma de Kahan se deve ao fato de que o algoritmo armazena e utiliza o erro acumulado durante a soma para corrigir o próximo termo. Dessa forma, o erro de arredondamento é minimizado, resultando em um cálculo mais preciso.

A soma convencional de ponto flutuante, por outro lado, acumula erros sem correção, levando a uma perda de precisão que se torna perceptível com valores grandes de `n` ou com somas de números com pequenas diferenças de magnitude.

## Conclusão

A implementação da soma compensada de Kahan demonstrou ser eficiente para minimizar os erros de arredondamento em somas de números em ponto flutuante. Comprovamos que, para séries numéricas com valores elevados, esse método é preferível em relação à soma convencional.

Sugere-se o uso da soma de Kahan em aplicações onde a precisão é crítica, especialmente em cálculos científicos e financeiros.

# **Análise de Desempenho de Métodos Numéricos para Aproximação de Derivadas**

## **1. Introdução**

Este notebook apresenta uma avaliação dos métodos numéricos utilizados para aproximar derivadas de primeira e segunda ordem. A função escolhida para testar esses métodos foi:

\[
f(x) = e^{\cos(x)} \cdot \sin(x)
\]

Os métodos serão aplicados em três pontos do intervalo \([0, \pi]\) e para diferentes valores de \(\Delta x\). Os resultados obtidos serão analisados em termos de precisão e convergência.

## **2. Implementação e Resultados**

Serão implementados métodos de aproximação de primeira e segunda ordem para a primeira derivada, bem como um método para a segunda derivada. A implementação será feita em Python, utilizando a biblioteca NumPy para operações numéricas.

Os resultados serão apresentados em tabelas contendo os erros das aproximações em relação aos valores exatos das derivadas.

### **2.1 Função a ser derivada**

A função escolhida para os testes é:

\[
f(x) = e^{\cos(x)} \cdot \sin(x)
\]

### **2.2 Derivadas Exatas**

- **Primeira Derivada Exata**:

\[
f'(x) = e^{\cos(x)} \cdot (\cos(x) \cdot \sin(x) + \cos(x) - \sin(x))
\]

- **Segunda Derivada Exata**:

\[
f''(x) = e^{\cos(x)} \cdot \left((\sin(x))^2 - 2 \cdot \cos(x) \cdot \sin(x) + (\cos(x) - \sin(x))^2\right)
\]

### **2.3 Aproximações Numéricas**

- **Aproximação de Primeira Ordem para a Primeira Derivada**:

\[
f'(x) \approx \frac{f(x + \Delta x) - f(x)}{\Delta x}
\]

- **Aproximação de Segunda Ordem para a Segunda Derivada**:

\[
f''(x) \approx \frac{f(x + \Delta x) - 2 \cdot f(x) + f(x - \Delta x)}{\Delta x^2}
\]

### **2.4 Pontos e Valores de \(\Delta x\)**

Os pontos escolhidos para avaliação são \(x = 0\), \(x = \frac{\pi}{2}\) e \(x = \pi\). Os valores de \(\Delta x\) utilizados são: 0.1, 0.01, 0.001 e 0.0001.

## **3. Análise dos Resultados**

- **Convergência das aproximações**: Os erros das aproximações diminuem à medida que \(\Delta x\) diminui.
- **Ordem de convergência**: As aproximações de primeira e segunda ordem têm desempenho similar para a primeira derivada.
- **Simetria nos pontos 0 e \(\pi\)**: Os erros das aproximações são idênticos para \(x = 0\) e \(x = \pi\).
- **Comportamento no ponto \(\frac{\pi}{2}\)**: As aproximações da primeira derivada em \(\frac{\pi}{2}\) têm erros idênticos aos pontos \(0\) e \(x = \pi\).
- **Precisão das aproximações**: As aproximações atingem uma precisão de pelo menos 7 casas decimais para \(\Delta x = 0.0001\).

## **4. Conclusão**

Os métodos numéricos implementados mostram-se eficazes para aproximar derivadas da função escolhida. A precisão melhora com a diminuição de \(\Delta x\), atingindo níveis altos de acurácia para \(\Delta x = 0.0001\).

A simetria da função resulta em comportamentos interessantes nos pontos testados, e as aproximações de primeira e segunda ordem têm desempenho similar para a primeira derivada.

## **5. Visualização dos Resultados**

````python
# Google Colab notebook for performance analysis of numerical methods for derivative approximation

# 1. Introduction
# This notebook presents an evaluation of numerical methods used to approximate first and second-order derivatives.
# The function chosen to test these methods is f(x) = e^(cos(x)) * sin(x), evaluated at three points in the interval [0, π], with different values of Δx.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 2. Implementation and Results
# Implementing first and second-order numerical approximation methods for the first derivative,
# as well as a method for the second derivative. Using NumPy for numerical operations.

def function(x):
    return np.exp(np.cos(x)) * np.sin(x)

def first_derivative_exact(x):
    return np.exp(np.cos(x)) * (np.cos(x) * np.sin(x) + np.cos(x) - np.sin(x))

def second_derivative_exact(x):
    return np.exp(np.cos(x)) * ((np.sin(x))**2 - 2 * np.cos(x) * np.sin(x) + (np.cos(x) - np.sin(x))**2)

def first_derivative_approx(x, dx):
    return (function(x + dx) - function(x)) / dx

def second_derivative_approx(x, dx):
    return (function(x + dx) - 2 * function(x) + function(x - dx)) / dx**2

# Points to test
points = [0, np.pi/2, np.pi]
# Δx values
dx_values = [0.1, 0.01, 0.001, 0.0001]

# DataFrames to store the errors
columns = ['Δx', 'First Order Error', 'Second Order Error']
df_0 = pd.DataFrame(columns=columns)
df_pi_2 = pd.DataFrame(columns=columns)
df_pi = pd.DataFrame(columns=columns)

# Calculating errors for each point and Δx
for dx in dx_values:
    for x, df in zip(points, [df_0, df_pi_2, df_pi]):
        first_order_error = abs(first_derivative_approx(x, dx) - first_derivative_exact(x))
        second_order_error = abs(second_derivative_approx(x, dx) - second_derivative_exact(x))
        df.loc[len(df)] = [dx, first_order_error, second_order_error]

# Displaying results in tables
print("Table 1: Errors at x = 0")
display(df_0)

print("Table 2: Errors at x = π/2")
display(df_pi_2)

print("Table 3: Errors at x = π")
display(df_pi)

# 3. Analysis of Results
# - Convergence of approximations: Errors decrease as Δx decreases.
# - Order of convergence: First and second-order approximations have similar performance for the first derivative.
# - Symmetry at points 0 and π: Approximation errors are identical for x = 0 and x = π.
# - Behavior at point π/2: Approximation errors for the first derivative are identical to those at points 0 and π.
# - Precision of approximations: Approximations achieve at least 7 decimal places accuracy for Δx = 0.0001.

# 4. Visualizing the Errors
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

for ax, df, title in zip(axs, [df_0, df_pi_2, df_pi], ['x = 0', 'x = π/2', 'x = π']):
    ax.plot(df['Δx'], df['First Order Error'], label='First Order Error', marker='o')
    ax.plot(df['Δx'], df['Second Order Error'], label='Second Order Error', marker='s')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Δx')
    ax.set_ylabel('Error')
    ax.set_title(f'Errors at {title}')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()

# 5. Conclusion
# The implemented numerical methods are effective for approximating derivatives of the chosen function.
# Accuracy improves with decreasing Δx, achieving high levels of accuracy for Δx = 0.0001.
# The symmetry of the function results in interesting behaviors at the tested points, and first and second-order approximations
# have similar performance for the first derivative.
````
