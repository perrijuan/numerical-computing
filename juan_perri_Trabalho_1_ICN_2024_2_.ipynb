
## Tarefa #2: Soma compensada de Kahan

Este é um algoritmo de tipo geral usado para calcular a soma de um conjunto de números $x_1,x_2\dots,x_n$, o objetivo  do mesmo é diminuir o efeito dos erros de arredondamento do computador no resultado da soma. O algoritmo foi desenvolvido pelo matemático canadense [Willian Kahan](https://pt.wikipedia.org/wiki/William_Kahan).

```python
# Adicione aqui seus códigos, mostrando os resultados pedidos

#possui erro de precisao

def sum(x):
  sum = 0
  for i in x:
    sum = sum + i
  return sum


entrada = list(map(float, input().split()))
resultado = sum(entrada)

print(f"Resultado do somatório de Kahan: {resultado:.16f}")
```

```python
def sum_kahan(x):
  sum = 0.0
  c = 0.0
  for i in x:
    y = i - c
    t = sum + y
    c = (t - sum) - y
    sum = t
  return sum
#fazendo uma lista com os valores em float para calcular e evitar o erro
entrada = list(map(float,input().split()))

resultado = sum_kahan(entrada)
#usando uma alta precisao para verificar o erro
print(f"Resultado do somatório de Kahan: {resultado:.16f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple, List, Dict
from dataclasses import dataclass

@dataclass
class AnaliseErros:
    erro_maximo: float
    erro_medio: float
    erro_acumulado: float
    erro_relativo: float
    valor_esperado: float
    valor_calculado: float

# Funções de soma modificadas para retornar apenas o valor final
def sum_kahan_final(x):
    total = 0.0
    c = 0.0
    for i in x:
        y = i - c
        t = total + y
        c = (t - total) - y
        total = t
    return total

def sum_conventional_final(x):
    total = 0.0
    for i in x:
        total += i
    return total

# Funções de soma que retornam valores parciais (para visualização)
def sum_kahan_parcial(x):
    total = 0.0
    c = 0.0
    soma_parcial = []
    for i in x:
        y = i - c
        t = total + y
        c = (t - total) - y
        total = t
        soma_parcial.append(total)
    return np.array(soma_parcial)

def sum_conventional_parcial(x):
    total = 0.0
    soma_parcial = []
    for i in x:
        total += i
        soma_parcial.append(total)
    return np.array(soma_parcial)

def analisar_erros(valores: np.ndarray, func_soma, valor_esperado: float) -> AnaliseErros:
    """
    Analisa diferentes tipos de erros para uma função de soma
    """
    resultado = func_soma(valores)
    erro_abs = abs(resultado - valor_esperado)
    erro_rel = erro_abs / abs(valor_esperado) if valor_esperado != 0 else 0

    # Calcula somas parciais para análise de erro acumulado
    somas_parciais = sum_conventional_parcial(valores)
    valores_esperados = np.arange(1, len(valores) + 1) * valores[0]
    erro_acumulado = np.sum(np.abs(somas_parciais - valores_esperados))

    return AnaliseErros(
        erro_maximo=erro_abs,
        erro_medio=erro_abs / len(valores),
        erro_acumulado=erro_acumulado,
        erro_relativo=erro_rel,
        valor_esperado=valor_esperado,
        valor_calculado=resultado
    )

def comparar_metodos(n_elementos: List[int], magnitude: float) -> Dict[str, List[AnaliseErros]]:
    """
    Compara os métodos de soma para diferentes tamanhos de entrada
    """
    resultados = {'Convencional': [], 'Kahan': []}

    for n in n_elementos:
        # Criar array com valores pequenos
        valores = np.ones(n) * magnitude
        valor_esperado = n * magnitude

        # Análise para soma convencional
        resultados['Convencional'].append(analisar_erros(valores, sum_conventional_final, valor_esperado))

        # Análise para soma de Kahan
        resultados['Kahan'].append(analisar_erros(valores, sum_kahan_final, valor_esperado))

    return resultados

def plotar_comparacao(n_elementos: List[int], resultados: Dict[str, List[AnaliseErros]]):
    """
    Gera gráficos comparativos dos diferentes tipos de erro
    """
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

    cores = {'Convencional': '#6a0dad', 'Kahan': '#1e90ff'}

    for metodo in resultados:
        erros_max = [r.erro_maximo for r in resultados[metodo]]
        erros_med = [r.erro_medio for r in resultados[metodo]]
        erros_acum = [r.erro_acumulado for r in resultados[metodo]]
        erros_rel = [r.erro_relativo for r in resultados[metodo]]

        # Erro Máximo
        ax1.plot(n_elementos, erros_max, 'o-', label=metodo, color=cores[metodo])
        ax1.set_title('Erro Máximo')
        ax1.set_yscale('log')
        ax1.grid(True)

        # Erro Médio
        ax2.plot(n_elementos, erros_med, 'o-', label=metodo, color=cores[metodo])
        ax2.set_title('Erro Médio')
        ax2.set_yscale('log')
        ax2.grid(True)

        # Erro Acumulado
        ax3.plot(n_elementos, erros_acum, 'o-', label=metodo, color=cores[metodo])
        ax3.set_title('Erro Acumulado')
        ax3.set_yscale('log')
        ax3.grid(True)

        # Erro Relativo
        ax4.plot(n_elementos, erros_rel, 'o-', label=metodo, color=cores[metodo])
        ax4.set_title('Erro Relativo')
        ax4.set_yscale('log')
        ax4.grid(True)

    for ax in [ax1, ax2, ax3, ax4]:
        ax.legend()
        ax.set_xlabel('Número de Elementos')
        ax.set_ylabel('Erro')

    plt.tight_layout()
    plt.show()

# Exemplo de uso
if __name__ == "__main__":
    # Definir tamanhos de entrada
    n_elementos = [100, 1000, 10000, 100000]

    # Magnitude dos valores (pequena para evidenciar erros de truncamento)
    magnitude = 1e-16

    # Realizar comparação
    resultados = comparar_metodos(n_elementos, magnitude)

    # Plotar resultados
    plotar_comparacao(n_elementos, resultados)

    # Imprimir resultados detalhados
    print("\nResultados Detalhados:")
    for n, (conv, kahan) in zip(n_elementos, zip(resultados['Convencional'], resultados['Kahan'])):
        print(f"\nNúmero de elementos: {n}")
        print(f"{'Métrica':<15} {'Convencional':<15} {'Kahan':<15} {'Diferença':<15}")
        print("-" * 60)
        print(f"{'Erro Máximo':<15} {conv.erro_maximo:e} {kahan.erro_maximo:e} {conv.erro_maximo - kahan.erro_maximo:e}")
        print(f"{'Erro Médio':<15} {conv.erro_medio:e} {kahan.erro_medio:e} {conv.erro_medio - kahan.erro_medio:e}")
        print(f"{'Erro Acumulado':<15} {conv.erro_acumulado:e} {kahan.erro_acumulado:e} {conv.erro_acumulado - kahan.erro_acumulado:e}")
        print(f"{'Erro Relativo':<15} {conv.erro_relativo:e} {kahan.erro_relativo:e} {conv.erro_relativo - kahan.erro_relativo:e}")
```

```python
#letra a

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def primeira_derivada_forward(fj, dx):
    """
    Aproximação forward para primeira derivada
    fj = [f0, f1, f2] onde:
    f0 = f(x0)
    f1 = f(x0 + dx)
    f2 = f(x0 + 2dx)
    """
    return (-fj[2] + 4*fj[1] - 3*fj[0]) / (2*dx)

def primeira_derivada_central(fj, dx):
    """
    Aproximação central para primeira derivada
    fj = [f_neg1, f0, f1] onde:
    f_neg1 = f(x0 - dx)
    f0 = f(x0)
    f1 = f(x0 + dx)
    """
    return (fj[2] - fj[0]) / (2*dx)

def segunda_derivada_central(fj, dx):
    """
    Aproximação central para segunda derivada
    fj = [f_neg1, f0, f1] onde:
    f_neg1 = f(x0 - dx)
    f0 = f(x0)
    f1 = f(x0 + dx)
    """
    return (fj[2] - 2*fj[1] + fj[0]) / (dx**2)

# Função exemplo f(x) = e^(cos(x))sin(x)
def f(x):
    return np.exp(np.cos(x)) * np.sin(x)

# Pontos de análise
x_pontos = [0, np.pi/2, np.pi]
deltas = [0.1, 0.01, 0.001]

resultados = []

for x0 in x_pontos:
    for dx in deltas:
        # Gerando os valores fj para cada método
        fj_forward = [f(x0), f(x0 + dx), f(x0 + 2*dx)]
        fj_central = [f(x0 - dx), f(x0), f(x0 + dx)]

        # Calculando as aproximações
        d1_forward = primeira_derivada_forward(fj_forward, dx)
        d1_central = primeira_derivada_central(fj_central, dx)
        d2_central = segunda_derivada_central(fj_central, dx)

        # Armazenando resultados
        resultados.append({
            'x': f'{x0:.2f}',
            'Δx': dx,
            'Primeira Derivada (Forward)': d1_forward,
            'Primeira Derivada (Central)': d1_central,
            'Segunda Derivada (Central)': d2_central
        })

# Criando DataFrame
df = pd.DataFrame(resultados)

# Visualização dos resultados
plt.figure(figsize=(15, 5))
for i, metodo in enumerate(['Primeira Derivada (Forward)',
                          'Primeira Derivada (Central)',
                          'Segunda Derivada (Central)']):
    plt.subplot(1, 3, i+1)
    for dx in deltas:
        data = df[df['Δx'] == dx]
        plt.plot(data['x'], data[metodo], 'o-', label=f'Δx = {dx}')
    plt.title(metodo)
    plt.xlabel('x')
    plt.ylabel('Valor')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

# Exibindo tabela de resultados
pd.set_option('display.float_format', lambda x: '%.6f' % x)
print("\nTabela de Resultados:")
print(df)

# Salvando resultados
df.to_csv('resultados_derivadas.csv', index=False)
```

```python
#letra b

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def f(x):
    return np.exp(np.cos(x)) * np.sin(x)

def primeira_derivada_forward(f, x, h):
    return (-f(x + 2*h) + 4*f(x + h) - 3*f(x)) / (2*h)

def primeira_derivada_central(f, x, h):
    return (f(x + h) - f(x - h)) / (2*h)

def segunda_derivada_forward(f, x, h):
    return (-f(x + 3*h) + 4*f(x + 2*h) - 5*f(x + h) + 2*f(x)) / (h**2)

def segunda_derivada_central(f, x, h):
    return (f(x + h) - 2*f(x) + f(x - h)) / (h**2)

def calcular_derivadas(x_ponto, delta_x):
    # Calcular aproximações
    d1_forward = primeira_derivada_forward(f, x_ponto, delta_x)
    d1_central = primeira_derivada_central(f, x_ponto, delta_x)
    d2_forward = segunda_derivada_forward(f, x_ponto, delta_x)
    d2_central = segunda_derivada_central(f, x_ponto, delta_x)

    return {
        'Primeira Derivada (Forward)': d1_forward,
        'Primeira Derivada (Central)': d1_central,
        'Segunda Derivada (Forward)': d2_forward,
        'Segunda Derivada (Central)': d2_central
    }

# Pontos de análise e deltas
x_pontos = [0, np.pi/2, np.pi]
deltas = [0.1, 0.01, 0.001, 0.0001]

# Criar DataFrame para armazenar resultados
resultados = []

for x in x_pontos:
    for delta in deltas:
        derivadas = calcular_derivadas(x, delta)
        for metodo, valor in derivadas.items():
            resultados.append({
                'Ponto x': f'{x:.2f}',
                'Δx': delta,
                'Método': metodo,
                'Valor': valor
            })

df = pd.DataFrame(resultados)

# Criar visualização
plt.figure(figsize=(15, 10))
for i, delta in enumerate(deltas):
    plt.subplot(2, 2, i+1)
    data = df[df['Δx'] == delta]
    sns.barplot(data=data, x='Ponto x', y='Valor', hue='Método')
    plt.title(f'Derivadas com Δx = {delta}')
    plt.xticks(rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

# Exibir tabela de resultados formatada
pd.set_option('display.float_format', lambda x: '%.6f' % x)
print("\nTabela de Resultados:")
print(df.to_string(index=False))

# Salvar resultados em um arquivo CSV
df.to_csv('resultados_derivadas.csv', index=False)
```

```python
#letra c

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def f(x):
    return np.exp(np.cos(x)) * np.sin(x)

def d1f_exact(x):
    """Primeira derivada exata de f(x) = e^(cos(x))sin(x)"""
    return np.exp(np.cos(x)) * (np.sin(x) * (-np.sin(x)) + np.cos(x))

def d2f_exact(x):
    """Segunda derivada exata de f(x) = e^(cos(x))sin(x)"""
    return np.exp(np.cos(x)) * (
        ((-np.sin(x))**2 + np.cos(x)) * (-np.sin(x)) +
        (-np.cos(x)) * (-np.sin(x)) +
        (np.sin(x) * (-np.sin(x)) + np.cos(x)) * (-np.sin(x)) +
        (-np.sin(x))
    )

def primeira_derivada_forward(fj, dx):
    """Aproximação forward para primeira derivada"""
    return (-fj[2] + 4*fj[1] - 3*fj[0]) / (2*dx)

def primeira_derivada_central(fj, dx):
    """Aproximação central para primeira derivada"""
    return (fj[2] - fj[0]) / (2*dx)

def segunda_derivada_central(fj, dx):
    """Aproximação central para segunda derivada"""
    return (fj[2] - 2*fj[1] + fj[0]) / (dx**2)

# Pontos de teste e deltas
x_pontos = [0, np.pi/2, np.pi]
deltas = [0.1, 0.01, 0.001, 0.0001]

resultados = []

for x0 in x_pontos:
    for dx in deltas:
        # Calculando valores da função nos pontos necessários
        fj_forward = [f(x0), f(x0 + dx), f(x0 + 2*dx)]
        fj_central = [f(x0 - dx), f(x0), f(x0 + dx)]

        # Calculando aproximações
        d1_forward = primeira_derivada_forward(fj_forward, dx)
        d1_central = primeira_derivada_central(fj_central, dx)
        d2_central = segunda_derivada_central(fj_central, dx)

        # Calculando valores exatos
        d1_exato = d1f_exact(x0)
        d2_exato = d2f_exact(x0)

        # Calculando erros absolutos
        erro_d1_forward = abs(d1_forward - d1_exato)
        erro_d1_central = abs(d1_central - d1_exato)
        erro_d2_central = abs(d2_central - d2_exato)

        resultados.append({
            'x': x0,
            'Δx': dx,
            'D1 Forward': d1_forward,
            'D1 Central': d1_central,
            'D2 Central': d2_central,
            'D1 Exato': d1_exato,
            'D2 Exato': d2_exato,
            'Erro D1 Forward': erro_d1_forward,
            'Erro D1 Central': erro_d1_central,
            'Erro D2 Central': erro_d2_central
        })

# Criando DataFrame
df = pd.DataFrame(resultados)

# Visualização dos erros
plt.figure(figsize=(15, 10))

# Plot para erros da primeira derivada
plt.subplot(2, 1, 1)
for x0 in x_pontos:
    data = df[df['x'] == x0]
    plt.plot(np.log10(data['Δx']), np.log10(data['Erro D1 Forward']), 'o-',
             label=f'Forward x={x0:.2f}')
    plt.plot(np.log10(data['Δx']), np.log10(data['Erro D1 Central']), 's--',
             label=f'Central x={x0:.2f}')
plt.title('Erros da Primeira Derivada (escala log)')
plt.xlabel('log10(Δx)')
plt.ylabel('log10(Erro)')
plt.legend()
plt.grid(True)

# Plot para erros da segunda derivada
plt.subplot(2, 1, 2)
for x0 in x_pontos:
    data = df[df['x'] == x0]
    plt.plot(np.log10(data['Δx']), np.log10(data['Erro D2 Central']), 'o-',
             label=f'x={x0:.2f}')
plt.title('Erros da Segunda Derivada (escala log)')
plt.xlabel('log10(Δx)')
plt.ylabel('log10(Erro)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Formatando e exibindo tabela de resultados
pd.set_option('display.float_format', lambda x: '%.8e' % x)
print("\nTabela de Resultados:")
print(df[['x', 'Δx', 'Erro D1 Forward', 'Erro D1 Central', 'Erro D2 Central']])
```
