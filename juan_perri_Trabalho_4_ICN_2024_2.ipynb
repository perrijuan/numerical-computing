```python
#item b

import numpy as np
import matplotlib.pyplot as plt

# Função que representa a derivada dy/dx
def funcao(x, y):
    return - y + np.sin(x)

# Método de Heun para resolver a PVI
def metodo_heun(x0, y0, n, intervalo):
    a, b = intervalo
    h = (b - a) / n  # Tamanho do passo

  #armazena os valores de x e y nos vetores
    x = x0
    y = y0
    y_list = [y]
    x_list = [x]

    for i in range(n):
        f_xy = funcao(x, y)
        y_p = y + h * f_xy  # Preditor (Método de Euler)
        x_new = x + h
        f_x_new = funcao(x_new, y_p)
        #metodo de heun (duas iclinacoes)
        y_new = y + (h / 2) * (f_xy + f_x_new)  # Corretor

        x = x_new
        y = y_new
        x_list.append(x)
        y_list.append(y)

        print(f"y{i+1} = {y}")

    return np.array(y_list), np.array(x_list)

# Entrada
x0 = 0.0           # Valor inicial de x
y0 = 1.0           # Valor inicial de y
n = 40          # Número de sub-intervalos
intervalo = [0, 10] # Intervalo [a, b]

y_vals, x_vals = metodo_heun(x0, y0, n, intervalo)
print("Valores de y:", y_vals)
print("Valores de x:", x_vals)

# Plotando um gráfico
plt.plot(x_vals, y_vals, marker='o', label='Aproximação de Heun')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Método de Heun para PVI')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x, y):
    return - y + np.sin(x)

def runge_kutta_4(x0, y0, funcao, x_final, n):
    h = (x_final - x0) / (n - 1)  # Calcula o passo h com base em n
    y = np.zeros(n)
    x = np.linspace(x0, x_final, n)
    y[0] = y0

    for i in range(n - 1):
        k1 = h * funcao(x[i], y[i])
        k2 = h * funcao(x[i] + h / 2, y[i] + k1 / 2)
        k3 = h * funcao(x[i] + h / 2, y[i] + k2 / 2)
        k4 = h * funcao(x[i] + h, y[i] + k3)
        y[i + 1] = y[i] + (k1 + 2 * k2 + 2 * k3 + k4) / 6

    return x, y

# Condição inicial
y0 = 1.0
x0 = 0.0
x_final = 10
n = 25  # Escolha o número de pontos desejado

x, y = runge_kutta_4(x0, y0, funcao, x_final, n)

# Exibindo os valores individuais de y
for i in range(len(y)):
    print(f"y{i} = {y[i]}")

# Exibindo os arrays completos
print("Valores de y:", y)
print("Valores de x:", x)

# Plotando o gráfico
plt.plot(x, y, marker='o', color='#8A2BE2', label='Solução Numérica')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Método de Runge-Kutta de 4ª Ordem para PVI')
plt.legend()
plt.grid(True)
plt.show()
```

b) Considere uma outra lei para a taxa de variação instantânea da velocidade do paraquedista dada pela EDO
$$\frac{dv}{dt}=g-\frac{C(v)}{m} v, $$
onde a função empiríca $C(v)$ que caracteriza a resistência do paraquedas é dada por
$$C(v) = c\left[1+\alpha \left(\frac{v}{v_{max}}\right)^\beta\right],$$
em que $c=12.5\,kg/s$, $\alpha=0.2$, $\beta =1.1$ e $v_{max} = 46\,m/s$.  (Observe que no caso $\alpha=0$ este modelo coincide com o apresentado no item a).)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def solucao_analitica(t, g=9.81, m=75, c=12.5):
    return (g*m/c) * (1 - np.exp(-(c/m)*t))

def parachutist_ode(t, v, g=9.81, m=75, c=12.5):
    return g - (c/m) * v

def metodo_numerico(t0, v0, n, t_final, metodo):
    h = t_final / n
    t = np.linspace(t0, t_final, n+1)
    v = np.zeros(n+1)
    v[0] = v0

    if metodo == 'euler':
        for i in range(n):
            v[i+1] = v[i] + h * parachutist_ode(t[i], v[i])

    elif metodo == 'heun':
        for i in range(n):
            k1 = parachutist_ode(t[i], v[i])
            k2 = parachutist_ode(t[i+1], v[i] + h*k1)
            v[i+1] = v[i] + (h/2) * (k1 + k2)

    elif metodo == 'rk4':
        for i in range(n):
            k1 = parachutist_ode(t[i], v[i])
            k2 = parachutist_ode(t[i] + h/2, v[i] + h*k1/2)
            k3 = parachutist_ode(t[i] + h/2, v[i] + h*k2/2)
            k4 = parachutist_ode(t[i+1], v[i] + h*k3)
            v[i+1] = v[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)

    return v[-1]



def criar_tabela_comparativa(t_final=25, erro_max=1e-4):
    metodos = ['euler', 'heun', 'rk4']
    passos_lista = [10, 20, 50, 100, 200]
    v_analitica = solucao_analitica(t_final)

    dados = []

    for metodo in metodos:
        for passos in passos_lista:
            v_numerica = metodo_numerico(0, 0, passos, t_final, metodo)
            erro = abs(v_numerica - v_analitica)
            erro_excedido = "Sim" if erro > erro_max else "Não"
            recomendacao = "Aumentar passos" if erro > erro_max else "OK"

            dados.append({
                'Método': metodo.upper(),
                'Número de Passos': passos,
                'Valor Numérico': round(v_numerica, 6),
                'Erro Absoluto': format(erro, '.2e'),
                'Erro Excedido': erro_excedido,
                'Recomendação': recomendacao
            })

    df = pd.DataFrame(dados)
    print(f"\nValor analítico: {v_analitica:.6f}")
    print(f"Erro máximo permitido: {erro_max}")
    return df

# Criar e exibir a tabela
df = criar_tabela_comparativa()
print("\nTabela comparativa dos métodos numéricos:")
print(df.to_string(index=False))

def criar_grafico_comparativo(t_final=25, erro_max=1e-4):
    metodos = ['euler', 'heun', 'rk4']
    # Usando mais pontos para ter uma visualização mais suave
    passos_lista = np.arange(10, 250, 10)
    v_analitica = solucao_analitica(t_final)

    plt.figure(figsize=(12, 8))

    for metodo in metodos:
        erros = []
        for passos in passos_lista:
            v_numerica = metodo_numerico(0, 0, passos, t_final, metodo)
            erro = abs(v_numerica - v_analitica)
            erros.append(erro)

        plt.plot(passos_lista, erros, marker='o', label=metodo.upper(), markersize=4)

    # Adicionar linha horizontal para o erro máximo
    plt.axhline(y=erro_max, color='r', linestyle='--', label='Erro Máximo Permitido')

    plt.xlabel('Número de Passos')
    plt.ylabel('Erro Absoluto')
    plt.title('Comparação do Erro Absoluto entre os Métodos Numéricos')
    plt.legend()
    plt.grid(True)
    plt.yscale('log')  # Escala logarítmica para melhor visualização
    plt.show()

# Criar o gráfico
criar_grafico_comparativo()
```

```python
# item b

import numpy as np
import matplotlib.pyplot as plt

def C(v, c=12.5, alpha=0.2, beta=1.1, v_max=46):
    """Função empírica C(v) que caracteriza a resistência do paraquedas"""
    return c * (1 + alpha * (v/v_max)**beta)

def parachutist_ode(t, v, g=9.81, m=75):
    """EDO com a nova função C(v)"""
    return g - (C(v) * v) / m

def solucao_analitica(t, g=9.81, m=75, c=12.5):
    """Solução analítica para o caso simples (alpha=0)"""
    return (g*m/c) * (1 - np.exp(-(c/m)*t))

def metodo_numerico(t0, v0, n, t_final, metodo):
    h = t_final / n
    t = np.linspace(t0, t_final, n+1)
    v = np.zeros(n+1)
    v[0] = v0

    if metodo == 'euler':
        for i in range(n):
            v[i+1] = v[i] + h * parachutist_ode(t[i], v[i])

    elif metodo == 'heun':
        for i in range(n):
            k1 = parachutist_ode(t[i], v[i])
            k2 = parachutist_ode(t[i+1], v[i] + h*k1)
            v[i+1] = v[i] + (h/2) * (k1 + k2)

    elif metodo == 'rk4':
        for i in range(n):
            k1 = parachutist_ode(t[i], v[i])
            k2 = parachutist_ode(t[i] + h/2, v[i] + h*k1/2)
            k3 = parachutist_ode(t[i] + h/2, v[i] + h*k2/2)
            k4 = parachutist_ode(t[i+1], v[i] + h*k3)
            v[i+1] = v[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)

    return t, v

# Condições iniciais e parâmetros
t0 = 0      # tempo inicial
v0 = 0      # velocidade inicial
t_final = 25  # tempo final
n = 200     # número de passos

# Calculando as soluções
t, v_euler = metodo_numerico(t0, v0, n, t_final, 'euler')
t, v_heun = metodo_numerico(t0, v0, n, t_final, 'heun')
t, v_rk4 = metodo_numerico(t0, v0, n, t_final, 'rk4')
v_analitica = solucao_analitica(t)

# Criando os três gráficos separados
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# Gráfico 1: Euler vs Analítica
ax1.plot(t, v_euler, 'b-', label='Euler')
ax1.plot(t, v_analitica, 'r--', label='Analítica')
ax1.set_xlabel('Tempo (s)')
ax1.set_ylabel('Velocidade (m/s)')
ax1.set_title('Método de Euler vs Solução Analítica')
ax1.grid(True)
ax1.legend()

# Gráfico 2: Heun vs Analítica
ax2.plot(t, v_heun, 'g-', label='Heun')
ax2.plot(t, v_analitica, 'r--', label='Analítica')
ax2.set_xlabel('Tempo (s)')
ax2.set_ylabel('Velocidade (m/s)')
ax2.set_title('Método de Heun vs Solução Analítica')
ax2.grid(True)
ax2.legend()

# Gráfico 3: RK4 vs Analítica
ax3.plot(t, v_rk4, 'm-', label='RK4')
ax3.plot(t, v_analitica, 'r--', label='Analítica')
ax3.set_xlabel('Tempo (s)')
ax3.set_ylabel('Velocidade (m/s)')
ax3.set_title('Método de Runge-Kutta 4 vs Solução Analítica')
ax3.grid(True)
ax3.legend()

# Ajustando o layout
plt.tight_layout()
plt.show()

# Imprimindo alguns valores para comparação
print("\nComparação de valores em pontos específicos:")
print("\nt (s) | Analítica | Euler | Heun | RK4")
print("-" * 50)
for i in range(0, n+1, 40):  # Mostra alguns pontos selecionados
    print(f"{t[i]:5.1f} | {v_analitica[i]:9.4f} | {v_euler[i]:5.4f} | {v_heun[i]:5.4f} | {v_rk4[i]:5.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def C_constante(v, c=12.5):
    """Função C constante"""
    return c

def C_variavel(v, c=12.5, alpha=0.2, beta=1.1, v_max=46):
    """Função C(v) variável"""
    return c * (1 + alpha * (v/v_max)**beta)

def edo_constante(t, v, g=9.81, m=75):
    """EDO com C constante"""
    return g - (C_constante(v) * v) / m

def edo_variavel(t, v, g=9.81, m=75):
    """EDO com C(v) variável"""
    return g - (C_variavel(v) * v) / m

def metodo_numerico(t0, v0, n, t_final, metodo, edo):
    h = t_final / n
    t = np.linspace(t0, t_final, n+1)
    v = np.zeros(n+1)
    v[0] = v0

    if metodo == 'euler':
        for i in range(n):
            v[i+1] = v[i] + h * edo(t[i], v[i])

    elif metodo == 'heun':
        for i in range(n):
            k1 = edo(t[i], v[i])
            k2 = edo(t[i+1], v[i] + h*k1)
            v[i+1] = v[i] + (h/2) * (k1 + k2)

    elif metodo == 'rk4':
        for i in range(n):
            k1 = edo(t[i], v[i])
            k2 = edo(t[i] + h/2, v[i] + h*k1/2)
            k3 = edo(t[i] + h/2, v[i] + h*k2/2)
            k4 = edo(t[i+1], v[i] + h*k3)
            v[i+1] = v[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)

    return t, v

# Condições iniciais e parâmetros
t0 = 0
v0 = 0
t_final = 20
n = 200

# Calculando as soluções para C constante
t, v_euler_const = metodo_numerico(t0, v0, n, t_final, 'euler', edo_constante)
t, v_heun_const = metodo_numerico(t0, v0, n, t_final, 'heun', edo_constante)
t, v_rk4_const = metodo_numerico(t0, v0, n, t_final, 'rk4', edo_constante)

# Calculando as soluções para C(v) variável
t, v_euler_var = metodo_numerico(t0, v0, n, t_final, 'euler', edo_variavel)
t, v_heun_var = metodo_numerico(t0, v0, n, t_final, 'heun', edo_variavel)
t, v_rk4_var = metodo_numerico(t0, v0, n, t_final, 'rk4', edo_variavel)

# Criando dois gráficos lado a lado
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico para C constante
ax1.plot(t, v_euler_const, 'b-', label='Euler')
ax1.plot(t, v_heun_const, 'g-', label='Heun')
ax1.plot(t, v_rk4_const, 'm-', label='RK4')
ax1.set_xlabel('Tempo (s)')
ax1.set_ylabel('Velocidade (m/s)')
ax1.set_title('Métodos Numéricos com C constante')
ax1.grid(True)
ax1.legend()

# Gráfico para C(v) variável
ax2.plot(t, v_euler_var, 'b-', label='Euler')
ax2.plot(t, v_heun_var, 'g-', label='Heun')
ax2.plot(t, v_rk4_var, 'm-', label='RK4')
ax2.set_xlabel('Tempo (s)')
ax2.set_ylabel('Velocidade (m/s)')
ax2.set_title('Métodos Numéricos com C(v) variável')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()

# Imprimindo tabela comparativa
print("\nComparação de valores em pontos específicos:")
print("\nt (s) | Euler (C) | Heun (C) | RK4 (C) | Euler C(v) | Heun C(v) | RK4 C(v)")
print("-" * 80)
for i in range(0, n+1, 20):  # Mostra alguns pontos selecionados
    print(f"{t[i]:5.1f} | {v_euler_const[i]:9.4f} | {v_heun_const[i]:8.4f} | "
          f"{v_rk4_const[i]:7.4f} | {v_euler_var[i]:9.4f} | {v_heun_var[i]:8.4f} | "
          f"{v_rk4_var[i]:7.4f}")

# Calculando e imprimindo os valores máximos
print("\nVelocidades máximas atingidas:")
print(f"C constante:")
print(f"  Euler: {np.max(v_euler_const):.4f} m/s")
print(f"  Heun:  {np.max(v_heun_const):.4f} m/s")
print(f"  RK4:   {np.max(v_rk4_const):.4f} m/s")
print(f"\nC(v) variável:")
print(f"  Euler: {np.max(v_euler_var):.4f} m/s")
print(f"  Heun:  {np.max(v_heun_var):.4f} m/s")
print(f"  RK4:   {np.max(v_rk4_var):.4f} m/s")
```

## Tarefa #3: Propagação de uma doença infecciosa

Para modelar a dinâmica de uma doença infecciosa em que os indivíduos se recuperam sem imunidade, é usado o seguinte sistema de duas EDOs conhecido como modelo SIS:

$$
\left\{
\begin{aligned}
\frac{d S}{dt} &= -\beta S I + \gamma I\\
\frac{d I}{dt} &= \beta S I - \gamma I
\end{aligned}
\right.
$$

onde $S$ e $I$ representam o número de indivíduos suscetíveis e infectados, respectivamente, $t$ o tempo,  $\beta$ a velocidade de contacto ou infecção da doença e $\gamma$ a taxa de recuperação média.

Considere uma cidade com uma população de $10000$ indivíduos todos suscetíveis. Se 10 indivíduos infectados entram na cidade no tempo $t = 0$, qual será a quantidade de indivíduos suscetíveis e infectados após $4$ semanas? Use os parâmetros $\beta = 0.0003/(\mathit{indivíduo\cdot dia})$ e $\gamma = 0.17/\mathit{dia}$.

Utilize o seu código do método de RK de $4^a$ ordem (tarefa #1, item c)) para responder a essa questão. Procure uma aproximação com erro estimado inferior a $0.05$. Explique.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes do modelo SIS
N = 10000          # População total
BETA = 0.0003      # Taxa de infecção por indivíduo por dia
GAMMA = 0.17       # Taxa de recuperação por dia
TOL = 0.05         # Tolerância desejada
T_FINAL = 28       # 4 semanas em dias
I0 = 10            # Infectados iniciais
S0 = N - I0        # Suscetíveis iniciais

def sis_model(y, t, params):
    """Sistema de EDOs do modelo SIS"""
    S, I = y
    beta, gamma = params
    dSdt = -beta * S * I + gamma * I
    dIdt = beta * S * I - gamma * I
    return np.array([dSdt, dIdt])

def rk4_step(f, y, t, h, params):
    """Um passo do método Runge-Kutta 4ª ordem"""
    k1 = h * f(y, t, params)
    k2 = h * f(y + k1/2, t + h/2, params)
    k3 = h * f(y + k2/2, t + h/2, params)
    k4 = h * f(y + k3, t + h, params)
    return y + (k1 + 2*k2 + 2*k3 + k4) / 6

def solve_sis_adaptive(f, y0, t_span, params, tol=TOL):
    """Resolve o sistema com passo adaptativo para controle de erro"""
    t = [t_span[0]]
    y = [y0]
    h = 0.1  # Passo inicial (em dias)

    while t[-1] < t_span[1]:
        # Previne ultrapassar tempo final
        if t[-1] + h > t_span[1]:
            h = t_span[1] - t[-1]

        # Solução com passo h
        y1 = rk4_step(f, y[-1], t[-1], h, params)

        # Solução com dois passos h/2
        y_meio = rk4_step(f, y[-1], t[-1], h/2, params)
        y2 = rk4_step(f, y_meio, t[-1] + h/2, h/2, params)

        # Estimativa do erro local
        erro = np.max(np.abs(y1 - y2))

        if erro < tol:
            t.append(t[-1] + h)
            y.append(y2)
            # Aumenta o passo se erro muito pequeno
            if erro < tol/10:
                h = min(h*1.5, 1.0)
        else:
            # Reduz o passo se erro muito grande
            h *= 0.5
            continue

    return np.array(t), np.array(y)

# Parâmetros e condições iniciais
params = [BETA, GAMMA]
y0 = np.array([S0, I0])
t_span = [0, T_FINAL]

# Resolve o sistema
t, sol = solve_sis_adaptive(sis_model, y0, t_span, params)

# Solução analítica para comparação
def sol_analitica(t):
    """Calcula a solução analítica do modelo SIS"""
    S = S0 * np.exp(-BETA * I0 * t / N) + (GAMMA / BETA) * (1 - np.exp(-BETA * I0 * t / N))
    I = I0 * np.exp(-GAMMA * t) + (BETA * I0 / (BETA - GAMMA)) * (
        np.exp(-GAMMA * t) - np.exp(-BETA * I0 * t / N))
    return S, I

# Calcula solução analítica nos pontos t
S_an, I_an = sol_analitica(t)

# Calcula erros
erro_S = np.abs(S_an - sol[:,0]) / S_an
erro_I = np.abs(I_an - sol[:,1]) / I_an

print("\nResultados da Simulação do Modelo SIS")
print("-" * 70)
print("Dia    Suscetíveis    Infectados    Sol. Analítica    Erro Local")
print("-" * 70)

for i in range(len(t)):
    print(f"{t[i]:4.1f}  {sol[i,0]:11.1f}  {sol[i,1]:11.1f}  {I_an[i]:14.1f}  {erro_I[i]:11.4e}")

print("\nAnálise dos Erros:")
print(f"Erro máximo: {np.max(erro_I):.4e}")
print(f"Erro médio: {np.mean(erro_I):.4e}")
print(f"Erro mínimo: {np.min(erro_I):.4e}")

# Visualização dos resultados
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(t, sol[:,0], 'b-', label='Suscetíveis (RK4)')
plt.plot(t, sol[:,1], 'r-', label='Infectados (RK4)')

plt.xlabel('Tempo (dias)')
plt.ylabel('Número de indivíduos')
plt.title('Dinâmica da Doença')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.semilogy(t, erro_I, 'k-', label='Erro Local')
plt.axhline(y=TOL, color='r', linestyle='--', label=f'Tolerância ({TOL})')
plt.xlabel('Tempo (dias)')
plt.ylabel('Erro (escala log)')
plt.title('Erro Local ao Longo do Tempo')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Contexto
Nesta questão, resolvemos a equação diferencial ordinária (EDO):

$$y'(x) = -y(x) + \sin(x)$$

utilizando os métodos de Euler, Heun e Runge-Kutta de 4ª ordem.

### Descrição dos Métodos

1. **Método de Euler**:  
   $$y_{i+1} = y_i + h \cdot f(x_i, y_i)$$
   onde $h$ é o tamanho do passo.

2. **Método de Heun**:  
   $$y_{i+1} = y_i + \frac{h}{2} \cdot (k_1 + k_2)$$
   onde:
   $$k_1 = f(x_i, y_i)$$
   $$k_2 = f(x_{i+1}, y_i + h \cdot k_1)$$

3. **Método de Runge-Kutta de 4ª Ordem (RK4)**:  
   $$y_{i+1} = y_i + \frac{h}{6} \cdot (k_1 + 2k_2 + 2k_3 + k_4)$$
   onde:
   $$k_1 = f(x_i, y_i)$$
   $$k_2 = f(x_i + \frac{h}{2}, y_i + \frac{h}{2} \cdot k_1)$$
   $$k_3 = f(x_i + \frac{h}{2}, y_i + \frac{h}{2} \cdot k_2)$$
   $$k_4 = f(x_{i+1}, y_i + h \cdot k_3)$$

A solução analítica é:
